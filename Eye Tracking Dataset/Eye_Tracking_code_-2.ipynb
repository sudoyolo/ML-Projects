```python
# import necessary libraries
import pandas as pd
import os
import glob


# use glob to get all the csv files
# in the folder
import numpy as np
finaldf = pd.DataFrame(columns = ['Pupil diameter', 'Shannon Entropy x', 'Shannon Entropy y','Fixation duration'])

path = r"C:\\Users\\BME-5\\Desktop\\ttdfile\\Tdddata - Copy" # use your path
csv_files = glob.glob(os.path.join(path, "*.csv"))
csv_files

# loop over the list of csv files
for f in csv_files:
      
    # read the csv file 
    df = pd.read_csv(f)

    j = df[(df.Sensor == 'Accelerometer')].index
    df=df.drop(j)
    i = df[(df.Sensor == 'Gyroscope')].index
    df=df.drop(i)
    df=df.reset_index(drop=True)
    updated_df=df.copy()

    
    updated_df=df.copy()
    updated_df
    updated_df['Gaze point X']=updated_df['Gaze point X'].fillna(updated_df['Gaze point X'].mean())
    updated_df['Gaze point Y']=updated_df['Gaze point Y'].fillna(updated_df['Gaze point Y'].mean())
    updated_df['Gaze point 3D X']=updated_df['Gaze point 3D X'].fillna(updated_df['Gaze point 3D X'].mean())
    updated_df['Gaze point 3D Y']=updated_df['Gaze point 3D Y'].fillna(updated_df['Gaze point 3D Y'].mean())
    updated_df['Gaze point 3D Z']=updated_df['Gaze point 3D Z'].fillna(updated_df['Gaze point 3D Z'].mean())
    updated_df['Gaze direction left X']=updated_df['Gaze direction left X'].fillna(updated_df['Gaze direction left X'].mean())
    updated_df['Gaze direction left Y']=updated_df['Gaze direction left Y'].fillna(updated_df['Gaze direction left Y'].mean())
    updated_df['Gaze direction left Z']=updated_df['Gaze direction left Z'].fillna(updated_df['Gaze direction left Z'].mean())
    updated_df['Gaze direction right X']=updated_df['Gaze direction right X'].fillna(updated_df['Gaze direction right X'].mean())
    updated_df['Gaze direction right Y']=updated_df['Gaze direction right Y'].fillna(updated_df['Gaze direction right Y'].mean())
    updated_df['Gaze direction right Z']=updated_df['Gaze direction right Z'].fillna(updated_df['Gaze direction right Z'].mean())
    updated_df['Pupil position left X']=updated_df['Pupil position left X'].fillna(updated_df['Pupil position left X'].mean())
    updated_df['Pupil position left Y']=updated_df['Pupil position left Y'].fillna(updated_df['Pupil position left Y'].mean())
    updated_df['Pupil position left Z']=updated_df['Pupil position left Z'].fillna(updated_df['Pupil position left Z'].mean())
    updated_df['Pupil position right X']=updated_df['Pupil position right X'].fillna(updated_df['Pupil position right X'].mean())
    updated_df['Pupil position right Y']=updated_df['Pupil position right Y'].fillna(updated_df['Pupil position right Z'].mean())
    updated_df['Pupil position right Z']=updated_df['Pupil position right Z'].fillna(updated_df['Pupil position right Z'].mean())
    updated_df['Pupil diameter left']=updated_df['Pupil diameter left'].fillna(updated_df['Pupil diameter left'].mean())
    updated_df['Pupil diameter right']=updated_df['Pupil diameter right'].fillna(updated_df['Pupil diameter right'].mean())
    updated_df['Recording media width']=updated_df['Recording media width'].fillna(updated_df['Recording media width'].mean())
    updated_df['Recording media height']=updated_df['Recording media height'].fillna(updated_df['Recording media height'].mean())
    updated_df['Fixation point X']=updated_df['Fixation point X'].fillna(updated_df['Fixation point X'].mean())
    updated_df['Fixation point Y']=updated_df['Fixation point Y'].fillna(updated_df['Fixation point Y'].mean())

    from sklearn.preprocessing import LabelEncoder
    le = LabelEncoder()
    updated_df['Validity left'] = le.fit_transform(updated_df['Validity left'])
    updated_df['Validity right'] = le.fit_transform(updated_df['Validity right'])
    
    index_names = updated_df[ updated_df['Eye movement type'] == 'Unclassified' ].index
    updated_df.drop(index_names, inplace = True)

    index_names = updated_df[ updated_df['Eye movement type'] == 'EyesNotFound' ].index
    updated_df.drop(index_names, inplace = True)

    updated_df = updated_df[updated_df.Sensor.notnull()]
    updated_df=updated_df.reset_index(drop=True)

    udf=updated_df.copy()
    index_names = udf[ udf['Eye movement type'] == 'Saccade' ].index
    udf.drop(index_names, inplace = True)

    column_name = "Gaze event duration"
    column_sum = df[column_name].sum()
    mean_fixation_duration=column_sum/18954

    #pupil diameter
    a=updated_df['Pupil diameter left'].mean()
    b=updated_df['Pupil diameter right'].mean()
    meanpupil=(a+b)/2

    import pandas as pd
    from scipy import stats
    a=updated_df['Gaze point X']
    b=updated_df['Gaze point Y']
    count1=a.value_counts()
    count2=b.value_counts()
    entropy_x = stats.entropy(count1)
    entropy_y=stats.entropy(count2)

    finaldf = finaldf.append({'Pupil diameter' :meanpupil , 'Shannon Entropy x' : entropy_x, 'Shannon Entropy y' : entropy_y, 'Fixation duration' : mean_fixation_duration}, 
                ignore_index = True)
```

```python
x= lastfinal[['Pupil diameter', 'Shannon Entropy x', 'Shannon Entropy y','Fixation duration']] 

y= lastfinal.iloc[:, 5].values  
print(x)
print(y)
```

```python
#LOGISTIC REGRESSION
from sklearn import metrics
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
pred = lr.predict(X_test)
print(metrics.accuracy_score(pred,y_test))
```

```python
#KNN
from sklearn import metrics
from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)


X_train, X_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
classifier.fit(X_train, y_train)

pred = classifier.predict(X_test)
print(metrics.accuracy_score(pred,y_test))
```
